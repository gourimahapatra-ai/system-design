## 1. Introduction to OOP

Object-Oriented Programming (OOP) is a programming paradigm that organizes code into **objects** containing both **data (attributes)** and **behavior (methods)**.

### Four Pillars of OOP

- **Encapsulation**: Bundling data and methods together  
- **Inheritance**: Creating new classes from existing ones  
- **Polymorphism**: Using a single interface for different data types  
- **Abstraction**: Hiding complex implementation details

```python
# Step 3: Adding Methods
class Dog:
    """A class representing a dog with methods"""
    
    def __init__(self, name, age, breed):
        self.name = name
        self.age = age
        self.breed = breed
    
    def bark(self):
        """Make the dog bark"""
        return f"{self.name} says Woof!"
    
    def get_info(self):
        """Return formatted information about the dog"""
        return f"{self.name} is a {self.age}-year-old {self.breed}"
    
    def have_birthday(self):
        """Increment the dog's age by 1"""
        self.age += 1
        return f"Happy Birthday {self.name}! Now {self.age} years old."

# Using methods
max_dog = Dog("Max", 2, "Golden Retriever")
print(max_dog.bark())           # Output: Max says Woof!
print(max_dog.get_info())       # Output: Max is a 2-year-old Golden Retriever
print(max_dog.have_birthday())  # Output: Happy Birthday Max! Now 3 years old.
```

```python
# 4. Constructors and Destructors
# Step 4: Understanding Special Methods
class BankAccount:
    """A class representing a bank account"""
    
    def __init__(self, account_holder, initial_balance=0):
        """
        Constructor: Initialize a new bank account
        
        Args:
            account_holder (str): Name of account holder
            initial_balance (float): Starting balance (default: 0)
        """
        self.account_holder = account_holder
        self.balance = initial_balance
        print(f"Account created for {account_holder}")
    
    def __del__(self):
        """
        Destructor: Called when object is about to be destroyed
        """
        print(f"Account for {self.account_holder} is being closed")
    
    def deposit(self, amount):
        """Add money to the account"""
        self.balance += amount
        return f"Deposited ${amount}. New balance: ${self.balance}"
    
    def withdraw(self, amount):
        """Remove money from the account"""
        if amount > self.balance:
            return "Insufficient funds"
        self.balance -= amount
        return f"Withdrew ${amount}. New balance: ${self.balance}"

    def addInterest(self):
        self.balance += self.balance * 0.05
        return f"Added interest. New balance: ${self.balance}"

# Usage
account = BankAccount("Alice", 1000)
print(account.deposit(500))
print(account.withdraw(200))
print(account.addInterest())
# Destructor is called when program ends or object is deleted
```

```python
# 5. Inheritance
# Step 5: Creating Child Classes
class Animal:
    """Base class for all animals"""
    
    def __init__(self, name, species):
        """
        Initialize an Animal
        
        Args:
            name (str): Animal's name
            species (str): Animal's species
        """
        self.name = name
        self.species = species
    
    def make_sound(self):
        """Generic sound method"""
        return "Some generic sound"
    
    def get_info(self):
        """Return basic information"""
        return f"{self.name} is a {self.species}"


class Dog(Animal):
    """Dog class inheriting from Animal"""
    
    def __init__(self, name, breed):
        """Initialize a Dog"""
        # Call parent class constructor
        super().__init__(name, "Dog")
        self.breed = breed
    
    def make_sound(self):
        """Override parent method"""
        return "Woof! Woof!"
    
    def fetch(self):
        """Dog-specific method"""
        return f"{self.name} is fetching the ball!"


class Cat(Animal):
    """Cat class inheriting from Animal"""
    
    def __init__(self, name, color):
        super().__init__(name, "Cat")
        self.color = color
    
    def make_sound(self):
        """Override parent method"""
        return "Meow!"
    
    def scratch(self):
        """Cat-specific method"""
        return f"{self.name} is scratching the furniture!"


# Usage
dog = Dog("Buddy", "Labrador")
cat = Cat("Whiskers", "Orange")

print(dog.get_info())      # Output: Buddy is a Dog
print(dog.make_sound())    # Output: Woof! Woof!
print(dog.fetch())         # Output: Buddy is fetching the ball!

print(cat.get_info())      # Output: Whiskers is a Cat
print(cat.make_sound())    # Output: Meow!
print(cat.scratch())       # Output: Whiskers is scratching the furniture!

# Documentation:

# Child classes inherit attributes and methods from parent classes
# super() calls methods from the parent class
# Child classes can override parent methods
# Child classes can add new methods
```

```python
#Step 6: Multiple Inheritance
class Flyable:
    """Mixin class for flying capability"""
    
    def fly(self):
        return f"{self.name} is flying!"


class Swimmable:
    """Mixin class for swimming capability"""
    
    def swim(self):
        return f"{self.name} is swimming!"


class Duck(Animal, Flyable, Swimmable):
    """Duck class with multiple inheritance"""
    
    def __init__(self, name):
        super().__init__(name, "Duck")
    
    def make_sound(self):
        return "Quack!"


# Usage
donald = Duck("Donald")
print(donald.make_sound())  # Output: Quack!
print(donald.fly())         # Output: Donald is flying!
print(donald.swim())        # Output: Donald is swimming!
```

```python
# Encapsulation
# Step 7: Public, Protected, and Private Members
class BankAccount:
    """Demonstrating encapsulation in Python"""
    
    def __init__(self, account_number, balance):
        self.account_number = account_number    # Public attribute
        self._balance = balance                  # Protected attribute (convention)
        self.__pin = "1234"                      # Private attribute (name mangling)
    
    # Public method
    def get_balance(self):
        """Public method to access balance"""
        return self._balance
    
    # Protected method (convention)
    def _calculate_interest(self, rate):
        """Protected method for internal use"""
        return self._balance * rate
    
    # Private method (name mangling)
    def __verify_pin(self, pin):
        """Private method for PIN verification"""
        return pin == self.__pin
    
    # Public method using private method
    def withdraw(self, amount, pin):
        """Withdraw money with PIN verification"""
        if not self.__verify_pin(pin):
            return "Invalid PIN"
        if amount > self._balance:
            return "Insufficient funds"
        self._balance -= amount
        return f"Withdrew ${amount}. Balance: ${self._balance}"
    
    # Property decorator (getter)
    @property
    def balance(self):
        """Read-only property for balance"""
        return self._balance


# Usage
account = BankAccount("12345", 1000)

# Public access
print(account.account_number)      # Works: 12345
print(account.get_balance())       # Works: 1000
print(account.balance)             # Works through property: 1000

# Protected access (still accessible but discouraged)
print(account._balance)            # Works but discouraged: 1000

# Private access (name mangled)
# print(account.__pin)             # AttributeError
# Actual access (not recommended):
print(account._BankAccount__pin)   # Works but defeats purpose: 1234

print(account.withdraw(100, "1234"))  # Output: Withdrew $100. Balance: $900

# Documentation:

# Public: No underscore prefix (accessible everywhere)
# Protected: Single underscore _ (convention, not enforced)
# Private: Double underscore __ (name mangling)
# @property decorator creates read-only attributes
```

```python
# 7. Polymorphism|``
# Step 8: Method Overriding and Duck Typing

from math import pi

class Shape:
    """Base class for shapes"""
    
    def area(self):
        """Calculate area - to be overridden"""
        raise NotImplementedError("Subclass must implement area()")
    
    def perimeter(self):
        """Calculate perimeter - to be overridden"""
        raise NotImplementedError("Subclass must implement perimeter()")


class Rectangle(Shape):
    """Rectangle class"""
    
    def __init__(self, width, height):
        self.width = width
        self.height = height
    
    def area(self):
        return self.width * self.height
    
    def perimeter(self):
        return 2 * (self.width + self.height)


class Circle(Shape):
    """Circle class"""
    
    def __init__(self, radius):
        self.radius = radius
    
    def area(self):
        return pi * self.radius ** 2
    
    def perimeter(self):
        return 2 * pi * self.radius


class Triangle(Shape):
    """Triangle class"""
    
    def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c
    
    def area(self):
        # Heron's formula
        s = (self.a + self.b + self.c) / 2
        return (s * (s - self.a) * (s - self.b) * (s - self.c)) ** 0.5
    
    def perimeter(self):
        return self.a + self.b + self.c


# Polymorphism in action
def print_shape_info(shape):
    """
    Function that works with any Shape object
    Demonstrates polymorphism
    """
    print(f"{shape.__class__.__name__}:")
    print(f"  Area: {shape.area():.2f}")
    print(f"  Perimeter: {shape.perimeter():.2f}")
    print()


# Usage
shapes = [
    Rectangle(5, 10),
    Circle(7),
    Triangle(3, 4, 5)
]

for shape in shapes:
    print_shape_info(shape)
```

```python
# Step 9: Operator Overloading
class Vector:
    """A 2D vector class with operator overloading"""
    
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    def __str__(self):
        """String representation for print()"""
        return f"Vector({self.x}, {self.y})"
    
    def __repr__(self):
        """Official string representation"""
        return f"Vector({self.x}, {self.y})"
    
    def __add__(self, other):
        """Overload + operator"""
        return Vector(self.x + other.x, self.y + other.y)
    
    def __sub__(self, other):
        """Overload - operator"""
        return Vector(self.x - other.x, self.y - other.y)
    
    def __mul__(self, scalar):
        """Overload * operator (scalar multiplication)"""
        return Vector(self.x * scalar, self.y * scalar)
    
    def __eq__(self, other):
        """Overload == operator"""
        return self.x == other.x and self.y == other.y
    
    def __len__(self):
        """Overload len() function"""
        return int((self.x**2 + self.y**2)**0.5)


# Usage
v1 = Vector(2, 3)
v2 = Vector(4, 1)

print(v1 + v2)          # Output: Vector(6, 4)
print(v1 - v2)          # Output: Vector(-2, 2)
print(v1 * 3)           # Output: Vector(6, 9)
print(v1 == v2)         # Output: False
print(len(v1))          # Output: 3
```

```python
# 8. Abstraction
# Step 10: Abstract Base Classes
from abc import ABC, abstractmethod

class Payment(ABC):
    """Abstract base class for payment methods"""
    
    def __init__(self, amount):
        self.amount = amount
    
    @abstractmethod
    def pay(self):
        """Process payment - must be implemented by subclasses"""
        pass
    
    @abstractmethod
    def refund(self):
        """Process refund - must be implemented by subclasses"""
        pass
    
    def print_receipt(self):
        """Common method for all payment types"""
        return f"Receipt for ${self.amount}"


class CreditCardPayment(Payment):
    """Credit card payment implementation"""
    
    def __init__(self, amount, card_number):
        super().__init__(amount)
        self.card_number = card_number
    
    def pay(self):
        return f"Paid ${self.amount} with credit card ending in {self.card_number[-4:]}"
    
    def refund(self):
        return f"Refunded ${self.amount} to credit card ending in {self.card_number[-4:]}"


class PayPalPayment(Payment):
    """PayPal payment implementation"""
    
    def __init__(self, amount, email):
        super().__init__(amount)
        self.email = email
    
    def pay(self):
        return f"Paid ${self.amount} via PayPal account {self.email}"
    
    def refund(self):
        return f"Refunded ${self.amount} to PayPal account {self.email}"


# Usage
payments = [
    CreditCardPayment(100, "1234567890123456"),
    PayPalPayment(50, "user@example.com")
]

for payment in payments:
    print(payment.pay())
    print(payment.print_receipt())
    print()

# This would raise an error (can't instantiate abstract class):
# payment = Payment(100)
```

```python
# 9. Advanced Concepts
# Step 11: Class Methods and Static Methods
class Employee:
    """Employee class demonstrating class and static methods"""
    
    # Class variable
    num_employees = 0
    raise_amount = 1.04
    
    def __init__(self, first, last, pay):
        self.first = first
        self.last = last
        self.pay = pay
        
        Employee.num_employees += 1
    
    # Instance method
    def fullname(self):
        return f"{self.first} {self.last}"
    
    def apply_raise(self):
        self.pay = int(self.pay * self.raise_amount)
    
    # Class method
    @classmethod
    def set_raise_amount(cls, amount):
        """Modify class variable"""
        cls.raise_amount = amount
    
    @classmethod
    def from_string(cls, emp_str):
        """Alternative constructor"""
        first, last, pay = emp_str.split('-')
        return cls(first, last, int(pay))
    
    # Static method
    @staticmethod
    def is_workday(day):
        """Utility method that doesn't depend on instance or class"""
        if day.weekday() == 5 or day.weekday() == 6:
            return False
        return True


# Usage
emp1 = Employee("John", "Doe", 50000)
emp2 = Employee("Jane", "Smith", 60000)

print(Employee.num_employees)  # Output: 2

# Class method usage
Employee.set_raise_amount(1.05)
print(Employee.raise_amount)   # Output: 1.05

# Alternative constructor
emp3 = Employee.from_string("Mike-Johnson-55000")
print(emp3.fullname())         # Output: Mike Johnson

# Static method usage
import datetime
my_date = datetime.date(2024, 7, 10)
print(Employee.is_workday(my_date))  # Output: True
```

```python
#Step 12: Property Decorators

class Temperature:
    """Temperature class with property decorators"""
    
    def __init__(self, celsius):
        self._celsius = celsius
    
    @property
    def celsius(self):
        """Get temperature in Celsius"""
        return self._celsius
    
    @celsius.setter
    def celsius(self, value):
        """Set temperature in Celsius"""
        if value < -273.15:
            raise ValueError("Temperature below absolute zero is not possible")
        self._celsius = value
    
    @property
    def fahrenheit(self):
        """Get temperature in Fahrenheit"""
        return (self._celsius * 9/5) + 32
    
    @fahrenheit.setter
    def fahrenheit(self, value):
        """Set temperature using Fahrenheit"""
        self.celsius = (value - 32) * 5/9
    
    @property
    def kelvin(self):
        """Get temperature in Kelvin"""
        return self._celsius + 273.15


# Usage
temp = Temperature(25)
print(f"Celsius: {temp.celsius}")       # Output: Celsius: 25
print(f"Fahrenheit: {temp.fahrenheit}") # Output: Fahrenheit: 77.0
print(f"Kelvin: {temp.kelvin}")         # Output: Kelvin: 298.15

temp.fahrenheit = 100
print(f"Celsius: {temp.celsius}")       # Output: Celsius: 37.77777777777778
```

```python
#Step 13: Magic Methods (Dunder Methods)
class Book:
    """Book class with various magic methods"""
    
    def __init__(self, title, author, pages):
        self.title = title
        self.author = author
        self.pages = pages
    
    def __str__(self):
        """Informal string representation"""
        return f"'{self.title}' by {self.author}"
    
    def __repr__(self):
        """Official string representation"""
        return f"Book('{self.title}', '{self.author}', {self.pages})"
    
    def __len__(self):
        """Return number of pages"""
        return self.pages
    
    def __eq__(self, other):
        """Check equality based on title and author"""
        return self.title == other.title and self.author == other.author
    
    def __lt__(self, other):
        """Compare books by page count"""
        return self.pages < other.pages
    
    def __add__(self, other):
        """Combine page counts"""
        return self.pages + other.pages
    
    def __contains__(self, keyword):
        """Check if keyword is in title"""
        return keyword.lower() in self.title.lower()


# Usage
book1 = Book("Python Crash Course", "Eric Matthes", 544)
book2 = Book("Clean Code", "Robert Martin", 464)

print(str(book1))               # Output: 'Python Crash Course' by Eric Matthes
print(repr(book1))              # Output: Book('Python Crash Course', 'Eric Matthes', 544)
print(len(book1))               # Output: 544
print(book1 == book2)           # Output: False
print(book1 > book2)            # Output: True
print(book1 + book2)            # Output: 1008
print("Python" in book1)        # Output: True
```

```python
# 10. Best Practices
# Guidelines for Clean OOP Code:

# Single Responsibility Principle: Each class should have one clear purpose
# Use Descriptive Names: Class names should be nouns, method names should be verbs
# Keep It Simple: Don't over-engineer; start simple and refactor as needed
# Composition Over Inheritance: Favor has-a relationships over is-a when appropriate
# Document Your Code: Use docstrings for classes and methods
# Follow PEP 8: Use Python's style guide for consistent formatting

# Complete Example Project: Library Management System
from datetime import datetime, timedelta
from typing import List, Optional

class Book:
    """Represents a book in the library"""
    
    def __init__(self, isbn: str, title: str, author: str, copies: int = 1):
        self.isbn = isbn
        self.title = title
        self.author = author
        self.total_copies = copies
        self.available_copies = copies
    
    def __str__(self):
        return f"{self.title} by {self.author} (ISBN: {self.isbn})"
    
    def is_available(self) -> bool:
        """Check if book is available for borrowing"""
        return self.available_copies > 0
    
    def borrow(self) -> bool:
        """Borrow a copy if available"""
        if self.is_available():
            self.available_copies -= 1
            return True
        return False
    
    def return_book(self) -> None:
        """Return a borrowed copy"""
        if self.available_copies < self.total_copies:
            self.available_copies += 1


class Member:
    """Represents a library member"""
    
    MAX_BOOKS = 3
    
    def __init__(self, member_id: str, name: str, email: str):
        self.member_id = member_id
        self.name = name
        self.email = email
        self.borrowed_books: List['Loan'] = []
    
    def __str__(self):
        return f"Member: {self.name} (ID: {self.member_id})"
    
    def can_borrow(self) -> bool:
        """Check if member can borrow more books"""
        return len(self.borrowed_books) < self.MAX_BOOKS
    
    def add_loan(self, loan: 'Loan') -> None:
        """Add a loan to member's record"""
        self.borrowed_books.append(loan)
    
    def remove_loan(self, loan: 'Loan') -> None:
        """Remove a loan from member's record"""
        if loan in self.borrowed_books:
            self.borrowed_books.remove(loan)


class Loan:
    """Represents a book loan"""
    
    LOAN_PERIOD_DAYS = 14
    
    def __init__(self, book: Book, member: Member):
        self.book = book
        self.member = member
        self.borrow_date = datetime.now()
        self.due_date = self.borrow_date + timedelta(days=self.LOAN_PERIOD_DAYS)
        self.return_date: Optional[datetime] = None
    
    def is_overdue(self) -> bool:
        """Check if loan is overdue"""
        if self.return_date:
            return False
        return datetime.now() > self.due_date
    
    def calculate_fine(self) -> float:
        """Calculate late fine ($1 per day)"""
        if not self.is_overdue():
            return 0.0
        days_late = (datetime.now() - self.due_date).days
        return days_late * 1.0
    
    def return_loan(self) -> float:
        """Return the book and calculate any fine"""
        self.return_date = datetime.now()
        self.book.return_book()
        return self.calculate_fine()


class Library:
    """Main library system"""
    
    def __init__(self, name: str):
        self.name = name
        self.books: List[Book] = []
        self.members: List[Member] = []
        self.loans: List[Loan] = []
    
    def add_book(self, book: Book) -> None:
        """Add a book to the library"""
        self.books.append(book)
        print(f"Added: {book}")
    
    def register_member(self, member: Member) -> None:
        """Register a new member"""
        self.members.append(member)
        print(f"Registered: {member}")
    
    def find_book(self, isbn: str) -> Optional[Book]:
        """Find a book by ISBN"""
        for book in self.books:
            if book.isbn == isbn:
                return book
        return None
    
    def find_member(self, member_id: str) -> Optional[Member]:
        """Find a member by ID"""
        for member in self.members:
            if member.member_id == member_id:
                return member
        return None
    
    def borrow_book(self, member_id: str, isbn: str) -> bool:
        """Process a book borrowing"""
        member = self.find_member(member_id)
        book = self.find_book(isbn)
        
        if not member:
            print("Member not found")
            return False
        
        if not book:
            print("Book not found")
            return False
        
        if not member.can_borrow():
            print(f"{member.name} has reached borrowing limit")
            return False
        
        if not book.is_available():
            print(f"'{book.title}' is not available")
            return False
        
        if book.borrow():
            loan = Loan(book, member)
            member.add_loan(loan)
            self.loans.append(loan)
            print(f"{member.name} borrowed '{book.title}'")
            print(f"Due date: {loan.due_date.strftime('%Y-%m-%d')}")
            return True
        
        return False
    
    def return_book(self, member_id: str, isbn: str) -> bool:
        """Process a book return"""
        member = self.find_member(member_id)
        book = self.find_book(isbn)
        
        if not member or not book:
            print("Member or book not found")
            return False
        
        for loan in member.borrowed_books:
            if loan.book.isbn == isbn:
                fine = loan.return_loan()
                member.remove_loan(loan)
                print(f"{member.name} returned '{book.title}'")
                if fine > 0:
                    print(f"Late fine: ${fine:.2f}")
                return True
        
        print("No active loan found for this book")
        return False
    
    def list_available_books(self) -> None:
        """Display all available books"""
        print(f"\n=== Available Books at {self.name} ===")
        for book in self.books:
            if book.is_available():
                print(f"{book} - {book.available_copies} copies available")


# Demo Usage
if __name__ == "__main__":
    # Create library
    library = Library("City Central Library")
    
    # Add books
    library.add_book(Book("978-0-13-468599-1", "Clean Code", "Robert Martin", 3))
    library.add_book(Book("978-0-13-235088-4", "Clean Architecture", "Robert Martin", 2))
    library.add_book(Book("978-1-59327-928-8", "Python Crash Course", "Eric Matthes", 5))
    
    # Register members
    library.register_member(Member("M001", "Alice Johnson", "alice@email.com"))
    library.register_member(Member("M002", "Bob Smith", "bob@email.com"))
    
    print("\n" + "="*50 + "\n")
    
    # Borrow books
    library.borrow_book("M001", "978-0-13-468599-1")
    library.borrow_book("M001", "978-1-59327-928-8")
    library.borrow_book("M002", "978-0-13-235088-4")
    
    print("\n" + "="*50 + "\n")
    
    # List available books
    library.list_available_books()
    
    print("\n" + "="*50 + "\n")
    
    # Return a book
    library.return_book("M001", "978-0-13-468599-1")
    
    print("\n" + "="*50 + "\n")
    
    # List available books again
    library.list_available_books()

# Summary
# This guide covered:

# Basics: Classes, objects, attributes, and methods
# Core Concepts: Constructors, inheritance, encapsulation
# Advanced Features: Polymorphism, abstraction, operator overloading
# Special Methods: Properties, class methods, static methods, magic methods
# Best Practices: Clean code principles and real-world examples

# Next Steps:

# Practice creating your own classes
# Implement design patterns (Singleton, Factory, Observer)
# Study SOLID principles
# Build complete projects using OOP
# Explore advanced topics like metaclasses and descriptors

# Resources:

# Official Python Documentation: https://docs.python.org/3/tutorial/classes.html
# PEP 8 Style Guide: https://pep8.org/
# Real Python OOP Tutorials: https://
```
